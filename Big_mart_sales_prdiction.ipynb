```python

# storing all the independent variables as X
X = data_train.drop('Item_Outlet_Sales', axis=1)

# storing the dependent variable as y
y = data_train['Item_Outlet_Sales']
```

```python
# importing the sequential model
from keras.models import Sequential
```

```python
# importing different layers from keras
from keras.layers import InputLayer, Dense
```

```python
X_train.shape[1]
```

```python
# defining input neurons
input_neurons = X_train.shape[1]
```

```python
# define number of output neurons
output_neurons = 1
```

```python
# define hidden layers and neuron in each layer
number_of_hidden_layers = 1
neuron_hidden_layer_1 = 30
neuron_hidden_layer_2 = 20
neuron_hidden_layer_3 = 10
neuron_hidden_layer_4 = 5

```

```python
# defining the architecture of the model
model = Sequential()
model.add(InputLayer(input_shape=(input_neurons,)))
model.add(Dense(units=neuron_hidden_layer_1, activation='relu'))
model.add(Dense(units=neuron_hidden_layer_2, activation='relu'))
model.add(Dense(units=neuron_hidden_layer_3, activation='relu'))
model.add(Dense(units=neuron_hidden_layer_4, activation='relu'))


model.add(Dense(units=output_neurons,))
```

```python
# summary of the model
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
```

```python
>>> history = model.fit(X_train, y_train, epochs=300)
```

```python
def plot_history(history, key):
  plt.plot(history.history[key])
  plt.plot(history.history['val_'+key])
  plt.xlabel("Epochs")
  plt.ylabel(key)
  plt.legend([key, 'val_'+key])
  plt.show()
# Plot the history
plot_history(history, 'mae')
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(prediction, y_test))
```
